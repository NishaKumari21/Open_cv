```python
import cv2
import numpy as np

# Create a black image of size 512x512 with 3 color channels (RGB)
img = np.zeros((512, 512, 3))

while True:
    # Display the image in a window named "window"
    cv2.imshow("window", img)
    
    # Wait for 1 millisecond and check if 'x' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break  # Exit the loop when 'x' is pressed

# Destroy all created OpenCV windows
cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np

# Create a black image of size 512x512 with 3 color channels (RGB)
img = np.zeros((512, 512, 3))

# Define the mouse callback function
def draw(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:  # Left mouse button clicked
        cv2.circle(img, center=(x, y), radius=50, color=(0, 0, 255), thickness=-1)

# Create a window and set the mouse callback
cv2.namedWindow('window')
cv2.setMouseCallback('window', draw)

# Display the window in a loop until 'x' is pressed
while True:
    cv2.imshow("window", img)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cv2.destroyAllWindows()
```
